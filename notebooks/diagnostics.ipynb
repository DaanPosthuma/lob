```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
def load_json(filename : str):
    with open(filename) as f:
        return json.load(f)
```

```python
diagnostics = {'ST_QQQ': load_json('../diagnostics/ST_QQQ.json'),
               'MT_QQQ': load_json('../diagnostics/MT_QQQ.json'),
               'MT_SPY': load_json('../diagnostics/MT_SPY.json'),
               'MT_AMD': load_json('../diagnostics/MT_AMD.json'),
               'MT_IWM': load_json('../diagnostics/MT_IWM.json')}
```

```python
def hist_lags(data):
    lags = np.array(data['lags'])
    plt.hist(lags[lags < 10000], bins=100)
```

```python
hist_lags(diagnostics['ST_QQQ'])
hist_lags(diagnostics['MT_QQQ'])
plt.show()
```
